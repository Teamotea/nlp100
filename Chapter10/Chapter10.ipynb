```python
def translate_ja_to_en(article_ja, is_split_into_words=False):
    tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
    tokenizer.src_lang = "ja_XX"

    encoded_ja = tokenizer(
        article_ja,
        return_tensors="tf",
        is_split_into_words=is_split_into_words,
        padding=True,
        # truncation=True,
        # max_length=16
    )
    generated_tokens = model.generate(**encoded_ja, forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"])
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

    return translation
```

```python
def translate_ja_to_en_bs(article_ja, num_beams, is_split_into_words):
    tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
    tokenizer.src_lang = "ja_XX"

    # num_beams=1 で early_stopping=True だと warning が出るため、num_beams=1 のときは False にする
    early_stopping = False if num_beams == 1 else True

    encoded_ja = tokenizer(
        article_ja,
        return_tensors="tf",
        is_split_into_words=is_split_into_words,
        padding=True,
        # truncation=True,
        # max_length=16
    )
    generated_tokens = model.generate(
        **encoded_ja,
        forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"],
        num_beams=num_beams,
        early_stopping=early_stopping
    )
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

    return translation
```

```python
def calculate_bleu(num_beams=1, is_split_into_words=True):
    predictions = []
    size = 5

    for i in tqdm(range(0, len(src_text_dev), size)):
        batch = src_text_dev[i:i+size]
        batch_preds = translate_ja_to_en_bs(batch, num_beams, is_split_into_words=is_split_into_words)
        predictions += batch_preds

    bleu = evaluate.load('bleu')
    results = bleu.compute(predictions=predictions, references=tgt_text_dev)

    return results['bleu']
```

```python
num_beams_choices = [1, 2, 3, 4, 5]  # GPU のメモリが溢れると思われるため、小さい値を指定
bleu_scores = [calculate_bleu(num_beams) for num_beams in num_beams_choices]
```
